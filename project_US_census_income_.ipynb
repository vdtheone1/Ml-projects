```python
# CHECK FOR PERCENTAGE OF OUTLIERS
for cols in list(continious):
    q1=df[cols].quantile(0.25)
    q3=df[cols].quantile(0.75)
    iqr=q3-q1
    upl=q3+(1.5*iqr)
    lpl=q1-(1.5*iqr)
    out=df[(df[cols]>upl)|(df[cols]<lpl)].shape[0]
    pec=out/df.shape[0]*100
    if pec>0:
        print('% of outlier present in ',cols,':',out/df.shape[0]*100)
```
